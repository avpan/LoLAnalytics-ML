```python
%matplotlib inline
import numpy as np
import pandas as pd
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.charts import HeatMap, output_file, show
from bokeh.palettes import YlOrRd3,YlOrRd4,YlOrRd5
from bokeh.models import Legend,BoxSelectTool,HoverTool
import matplotlib.pyplot as plt
output_notebook()
```

```python
feature_df = pd.read_csv("feature_importances_esports.csv",index_col=0)
cols = list(feature_df.columns)
cols.pop()
```

```python
feature_df
```

```python
feature_importances = pd.melt(feature_df,id_vars=['time'],value_vars=cols,var_name='Features',value_name='Importance')
```

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,resize"
hover = HoverTool(tooltips=[("(feature,time,importance)", "(@x, @y, @values)")]) 
hm = HeatMap(feature_importances,x='Features',y='time',values='Importance',stat=None,width=500,height=500,
             palette=YlOrRd5,legend='top_right',title='Feature Importance',tools=[TOOLS,hover])
```

```python
#output_file('heatmap.html')
show(hm);
```

```python
score_df = pd.read_csv('predictionaccuracy.csv',index_col=0)
TOOLS = "pan,wheel_zoom,box_zoom,reset,resize"
hover = HoverTool(tooltips=[("time", "$index"),("(time, accuracy) (x,y)", "(@x, @y)")]) 
p = figure(title = 'Prediction Accuracy',plot_width=500,plot_height=500,tools=[TOOLS,hover],x_axis_label='Time(minutes)',y_axis_label='Accuracy Score')
p.line(score_df['time'],score_df['accuracyScore'],line_width=2,color='mediumpurple')
p.circle(score_df['time'],score_df['accuracyScore'],color='indigo',size=5)
show(p);
```
